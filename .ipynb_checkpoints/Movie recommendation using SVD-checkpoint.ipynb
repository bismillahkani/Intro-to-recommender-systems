```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv('movies.csv')
movies_df.head()
```

```python
ratings_df = pd.read_csv('ratings_small.csv')
ratings_df.head()
```

```python
R_df = ratings_df.pivot(index = 'userId', columns = 'movieId', values = 'rating').fillna(0)
R_df.head()
```

```python
R = R_df.as_matrix()
user_rating_mean = np.mean(R, axis=1)
R_demeaned = R - user_rating_mean.reshape(-1,1)
```

```python
R_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 50)
```

```python
sigma = np.diag(sigma)
```

```python
U
```

```python
sigma
```

```python
all_user_predicted_ratings = np.dot(np.dot(U,sigma),Vt) + user_rating_mean.reshape(-1,1)
```

```python
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = R_df.columns)
preds_df.head()
```

```python
def recommend_movies(preds_df, userID, movies_df, original_ratings_df, num_recommendation=5):
    
    user_row_number = userID - 1
    
    sorted_user_predictions = preds_df.iloc[user_row_number].sort_values(ascending=False)
    
    user_data = original_ratings_df[original_ratings_df.userId == (userID)]
    
    user_full = (user_data.merge(movies_df, how='left', left_on='movieId', right_on='movieId').sort_values(['rating'],ascending=False))
    
    print("User {0} has already rated {1} movies".format(userID, user_full.shape[0]))
    print("Recommending highest {0} predicted ratings movies not alread rated".format(num_recommendation))
    
    recommendations = movies_df[~movies_df['movieId'].isin(user_full['movieId'])].merge(pd.DataFrame(sorted_user_predictions).reset_index(),how='left',left_on='movieId',right_on='movieId').rename(columns = {user_row_number : 'Predictions'}).sort_values('Predictions', ascending=False).iloc[:num_recommendation, :-1]
    
    return user_full, recommendations
    
```

```python
already_rated, predictions = recommend_movies(preds_df, 2, movies_df, ratings_df, 5)
```
